Demo of neural network using titanic dataset

```python
from data_prep import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_path = './Titanic_data/train.csv'
test_path = './Titanic_data/test.csv'
labels = ['Survived']

# Prepare the data
train_data, test_data, col_names = load_data_train_test(train_path, test_path)

# Drop unnecessary columns
train_data = train_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin','Embarked'])
test_data = test_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'])

# Encode non-numeric columns
train_data = encode_non_numeric(train_data)
test_data = encode_non_numeric(test_data)

# Fill missing values
train_data = fill_missing_values(train_data)
test_data = fill_missing_values(test_data)

number_of_features = train_data.shape[1] - 1  # Exclude the target variable
number_of_labels = 1  # Survived column

print(f"Number of features: {number_of_features}")
print(f"Number of labels: {number_of_labels}")
```

```python
train_data, train_mean, train_std = normalize_data_z_score(train_data, labels)
test_data, test_mean, test_std = normalize_data_z_score(test_data, labels)

x_train, y_train = split_reshape_data(train_data, labels)
x_test, y_test = split_reshape_data(test_data, labels)
```
